```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Define numeric and categorical pipelines
# numeric_cols = Pipeline([
#     ('scaler', StandardScaler())
# ])

# categorical_cols = Pipeline([
#     ('encoder', OneHotEncoder(drop='first', sparse_output=False)),
#     ('scaler', StandardScaler())
# ])

# # Define transformer directly
# transformer = ColumnTransformer([
#     ('numeric_cols', numeric_cols, [0, 4, 5, 6, 7, 8, 9]),
#     ('categorical_cols', categorical_cols, [1, 2])
# ], remainder='passthrough')

# transformer.set_output(transform='pandas')

numeric_cols = Pipeline([
    ('scaler' ,StandardScaler())
    ])

categorical_cols = Pipeline([
        ('encoder' , OneHotEncoder(drop='first',sparse_output=False)),
        ('scaler',StandardScaler())
    ])

transformer = ColumnTransformer([
        ('numeric_cols',numeric_cols,[0,4,5,6,7,8,9]),
        ('categorical_cols',categorical_cols,[1,2])
    ],remainder='passthrough')

transformer.set_output(transform='pandas')

# Fit the transformer
transformer.fit(X_train)

# Save it safely
import dill
with open('model.pkl', 'wb') as f:
    dill.dump(transformer, f)
```
